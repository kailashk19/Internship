# Q 4. Write a python program to scrap first 10 product details which include product name , price , Image URL from http://www.bewakoof.com/bestseller?sort=popular

```python
!pip install bs4
!pip install requests
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('http://www.bewakoof.com/bestseller?sort=popular')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
product_name = soup.find('div',class_="productNaming bkf-ellipsis")
product_name
```

```python
price = soup.find('span',class_="sellingFastBox")
price
```

```python
image = soup.find('div',class_="productCardImg false")
image
```

```python
products =[]


for i in soup.find_all('div',class_="productNaming bkf-ellipsis"):
    products.append(i.text)
    
    
products
```

```python
prices =[]


for i in soup.find_all('span',class_="sellingFastBox"):
    prices.append(i.text)
    
    
prices 
```

```python
images =[]


for i in soup.find_all('div',class_="productCardImg false"):
    images.append(i.text)
    
    
images 
```

# Q. 6 Please visit http://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/ and scrap=
a) Paper title

c)Author

```python
!pip install bs4
!pip install requests
```

```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.keaipublishing.com/en/journals/artificial-intelligence-in-agriculture/most-downloaded-articles/')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
paper_title = soup.find('h2',class_="h5 article-title")
paper_title
```

```python
date = soup.find('p',class_="article-date")
date
```

```python
author = soup.find('p',class_="article-authors")
author
```
